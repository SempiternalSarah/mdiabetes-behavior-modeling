```python
from sklearn.manifold import TSNE
from experiment import Experiment
from utils.behavior_data import BehaviorData
import torch
import numpy as np
from utils.state_data import StateData
import matplotlib.pyplot as plt


respond_perc = .5


## TRADITIONAL STATES

bd = BehaviorData(minw=2, 
                  maxw=29, 
                  include_state=True, 
                  include_pid=False, 
                  expanded_states = False, 
                  top_respond_perc=respond_perc)

data = bd.features.numpy()
labels = bd.labels.numpy()

indices = np.arange(len(labels))
zeros = indices[labels[:, 0] == 1]
ones = indices[labels[:, 1] == 1]
twos = indices[labels[:, 2] == 1]
threes = indices[labels[:, 3] == 1]

fig, subplots = plt.subplots(1, 5, figsize=(20, 6))
for idx, perps in enumerate([5, 15, 25, 50, 100]):
    tsne = TSNE(n_components = 2, init="random", perplexity=perps, n_iter=400, learning_rate="auto")
    Y = tsne.fit_transform(data)
    subplots[idx].scatter(Y[zeros, 0], Y[zeros, 1], c="red")
    subplots[idx].scatter(Y[ones, 0], Y[ones, 1], c="orange")
    subplots[idx].scatter(Y[twos, 0], Y[twos, 1], c="yellow")
    subplots[idx].scatter(Y[threes, 0], Y[threes, 1], c="green")
    subplots[idx].set_title(f"Perp = {perps}")
# plt.ylabel("No. Participants")
# plt.xlabel("Response %")
plt.show()
```
